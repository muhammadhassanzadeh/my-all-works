## Recursion

```python
#fibonacci sequence by search tree

def f(n):
    if n<=2:
        return 1
    else:
        return f(n-1)+f(n-2)
f(20)   ,f(38)  
```

```python
import time
  
def ttime(n):
    begin = time.time()
    f(n)
    time.sleep(1)
    end = time.time()
    return end - begin


ttime(20)  ,ttime(38)  ,ttime(40)
```

## memoization approach

```python
#fibonacci sequence by memoization and subproblems make it efficient
    
save={}
def df(n):
    if n in save:
        return save[n]
    if n<=2:
        f=1  
    else:
         f=df(n-1)+df(n-2)
    save[n]=f
    return save[n]
 
df(38) , df(380),df(1000)
```

```python
import time
  
def dtime(n):
    begin = time.time()
    df(n)
    time.sleep(1)
    end = time.time()
    return end - begin
dtime(38) ,dtime(100),dtime(1000)
```

## plot of time

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

xx=[]
y=[]
yy=[]
for i in range(20,39):
    xx.append(i)
    y.append(ttime(i))
    yy.append(dtime(i))


plt.xlabel("number of terms of sequence")
plt.ylabel("time lasted for computation")

plt.plot( xx, y, 'r--' ,label='exponentiail time')
plt.plot( xx,yy,'b--' ,label='polynomial time')
plt.title('comparing the recursive searching and dynamic searching')
plt.legend()
plt.show()
```
