3 wave solution by exponentioal functions for CDJKM equation:
###### $$u_{xxxxy} + 4u_{xxy}u_{x} + 2u_{xxx}u_{y} + 6u_{xy}u_{xx} -\alpha u_{yyy}-2\beta \frac{\partial}{\partial x }(\frac{ \partial}{\partial x}(\frac{\partial ^{\theta} u}{\partial t^{\theta}}))= 0$$
###### $u=u(x,y,t)$ $$0 < \theta \leq 1$$

### one wave solution:
##### $$u=\frac{a_0+a_1e^{k_1x+l_1y-w_1t}}{b_0+b_1e^{k_1x+l_1y-w_1t}}$$




```python
%matplotlib inline
import numpy as np
from sympy import *
import matplotlib.pyplot as plt
```

```python
x,y,t=symbols('x y t ', real=True) 
a0,a1,b0,b1,k1,l1,w1=symbols('a0 a1 b0 b1 k1 l1 w1', real=True)
u=Eq(Function('u')(x,y,t),(a0+a1*exp(k1*x+l1*y-w1*t))/(b0+b1*exp(k1*x+l1*y-w1*t)))
u.subs({a1:1})
```

```python

#alph,beta=symbols('A B')
eta=Symbol('E')
#Defining our partial differential equation given three param thet alph and beta
def equation(U,thet,alph,beta):
    return Eq(U.diff(x,x,x,x,y)+4*U.diff(x,x,y)*U.diff(x)+2*U.diff(x,x,x)*U.diff(y)+6*U.diff(x,y)*U.diff(x,x)-alph*U.diff(y,y,y)
            -2*beta*(U.diff(t)).diff(x,x),0)
equ=equation(u.rhs,.5,1,1)
```

```python
eta1=Symbol('et')
eq_u=equ.subs({exp(k1*x + l1*y - t*w1):eta1,exp(2*k1*x + 2*l1*y - 2*t*w1):eta1**2,exp(3*k1*x + 3*l1*y - 3*t*w1):eta1**3,
          exp(4*k1*x + 4*l1*y - 4*t*w1):eta1**4,exp(5*k1*x + 5*l1*y - 5*t*w1):eta1**5}).lhs.factor()
```

```python
numer=fraction(eq_u)[0]
numer
```

```python
expan=numer.expand()
expan
```

```python
colec=expan.collect(eta1,evaluate=False)
colec
```

```python
coeff=list(colec.values())
coeff
```

```python

dd=solve(coeff,[a1,w1])
dd
```

# two wave solution:
## $$u(x,y,t)=\frac{2(k_1c_1e^{k_1x+l_1y-w_1t}+k_2c_2e^{k_2x+l_2y-w_2t}+a_{12}(k_1+k_2)c_1e^{k_1x+l_1y-w_1t}c_2e^{k_2x+l_2y-w_2t})}{1+c_1e^{k_1x+l_1y-w_1t}+c_2e^{k_2x+l_2y-w_2t}+a_{12}c_1e^{k_1x+l_1y-w_1t}c_2e^{k_2x+l_2y-w_2t}}$$

```python
c1,c2,a12,k1,l1,w1,k2,l2,w2=symbols('c1 c2 a12 k1 l1 w1 k2 l2 w2', real=True)
ta1=c1*exp(k1*x+l1*y-w1*t)
ta2=c2*exp(k2*x+l2*y-w2*t)
uu=Eq(Function('u_1')(x,y,t),(2*(k1*ta1+k2*ta2+a12*(k1+k2)*ta1*ta2))/(1+ta1+ta2+a12*ta1*ta2))
uu
```

```python
eq_u=equation(uu.rhs,.5,1,1)
print(eq_u)
```

```python
eta1,eta2=symbols('et1 et2')
eq_u=eq_u.subs({exp(k1*x + l1*y - t*w1):eta1,exp(k2*x + l2*y - t*w2):eta2}).lhs.factor()
print(eq_u)
```
